```python
import matplotlib.pyplot as plt
nomes= ['Rafael', 'Maria', 'Ana']
idades = [10,20,15]


plt.plot(nomes,idades)
```

INSERINDO GRAFICO EM BARRAS

Utilizando o plt.figure para redimencionar o tamanho do grafico

plt.label para adicionar legenda

```python
plt.figure(figsize=(5,5))
plt.bar(nomes,idades)
plt.xlabel('nome')
plt.ylabel('idades')
```
